## Visualizing the data

```python
sns.set_theme()

def add_title(x, hue, ax=None, title=None):
    if title:
        if ax:
            ax.set_title(title)
        else:
            plt.title(title)
    else:
        if ax:
            ax.set_title(rf"Distribution of {x} $\backsim$ {hue}")
        else:
            plt.title(rf"Distribution of {x} $\backsim$ {hue}")

            
def plot_hist(data, x, hue=None, stat="count", ax=None, xlabel=None, title=None):
    sns.histplot(data=data, x=x, hue=hue, stat=stat, multiple="stack", ax=ax)
    if ax:
        ax.set_xlabel([xlabel, x][xlabel is None])
        ax.set_ylabel(["occurences", "frequency"][stat != "count"])
        add_title(x, hue, ax, title)
    else:
        plt.xlabel([xlabel, x][xlabel is None])
        plt.ylabel(["occurences", "frequency"][stat != "count"])
        add_title(x, hue, ax, title)
        plt.show();
```

### sex

```python
plot_hist(df, "sex", "target", stat="percent")
```

```python
fig, axes = plt.subplots(1, 2)
fig.figure.set_figwidth(16)
fig.figure.set_figheight(13)
plot_hist(df[df["sex"] == 0], x="chol", hue=None, ax=axes[0], title="Distribution of cholesterol in women")
plot_hist(df[df["sex"] == 1], x="chol", hue=None, ax=axes[1], title="Distribution of cholesterol in men")
fig.suptitle("Distribution of Cholesterol (mg/dl)");
```

## The 500+ cholesterole looks like a severe outlier (if not data typo?). A google research tells us that it is possible though.

### age

```python
plot_hist(data=df[df["sex"] == 0], x="age", hue="target", stat="percent", title="Distribution of age for female")
plot_hist(data=df[df["sex"] == 1], x="age", hue="target", stat="percent", title="Distribution of age for male")
```

## The female samples seem to be older than the males

```python
cols = list(df.columns)
cols.pop(cols.index("age"))
cols.pop(cols.index("sex"))
cols.pop(cols.index("target"));
```

```python
for col in cols:
    fig, axes = plt.subplots(1, 2)
    fig.figure.set_figwidth(16)
    fig.figure.set_figheight(13)
    fig.suptitle(f"Distribution of {col}")
    plot_hist(data=df[df["sex"] == 0], x=col, hue="target", stat="percent", xlabel=label_dict[col] 
              title=f"Distribution of {col} for female", ax=axes[0])
    
    plot_hist(data=df[df["sex"] == 1], x=col, hue="target", stat="percent",
              title=f"Distribution of {col} for male", ax=axes[1])
    
```

## The % of sick men seem to be quite high. There seems to be a bias from hospitalization. Young hospitalization means there probably is something very wrong, unlike for elders.
## This effect, while not as severe in women samples, is still present there as well.
